# Assignment passing

```python
def evenodd(x):
    if x%2==0:
        print("Even")
evenodd(6)
```

#Keyword argumments

```python
def student(fname,lname):
    print(fname,lname)

student(fname="practice",lname="daily")

student(lname="daily",fname="practice")
```

#Positional Argument

```python
def person(name,age):
    print("Hi i am",name)
    print("Hi i am ",age)
print("Case 1")
person("Ram",25)
print("case 2")
person(25,"Ram")
```

#Fuction within functions

```python
def f1():
    s="I love India"
    def f2():
        print(s)
    f2()
f1()        
```
